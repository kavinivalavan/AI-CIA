```python
import math

def alphabeta_pruning(depth, index, isMinimizing, node_values, max_depth, alpha, beta):
    if depth == max_depth:
        print(f"Leaf node reached at depth {depth}, returning value: {node_values[index]}")
        return node_values[index]

    if isMinimizing:
        optimal = math.inf
        print(f"Minimizer at depth {depth}")

        for i in range(2):
            value = alphabeta_pruning(depth + 1, index * 2 + i, False, node_values, max_depth, alpha, beta)
            print(f"Minimizer at depth {depth}, comparing value: {value} with optimal: {optimal}")
            optimal = min(optimal, value)
            beta = min(beta, optimal)

            if beta <= alpha:
                print(f"Minimizer at depth {depth}, pruning branches with beta: {beta} and alpha: {alpha}")
                break

        print(f"Minimizer at depth {depth}, selected optimal: {optimal}")
        return optimal
    else:
        optimal = -math.inf
        print(f"Maximizer at depth {depth}")

        for i in range(2):
            value = alphabeta_pruning(depth + 1, index * 2 + i, True, node_values, max_depth, alpha, beta)
            print(f"Maximizer at depth {depth}, comparing value: {value} with optimal: {optimal}")
            optimal = max(optimal, value)
            alpha = max(alpha, optimal)

            if beta <= alpha:
                print(f"Maximizer at depth {depth}, pruning branches with beta: {beta} and alpha: {alpha}")
                break

        print(f"Maximizer at depth {depth}, selected optimal: {optimal}")
        return optimal

def run_alphabeta(max_depth, node_values):
    alpha = -math.inf
    beta = math.inf
    
    print("For the minimizer:")
    result_min = alphabeta_pruning(0, 0, True, node_values, max_depth, alpha, beta)
    print(f"Optimal value for minimizer: {result_min}\n")

    print("For the maximizer:")
    alpha = -math.inf
    beta = math.inf
    result_max = alphabeta_pruning(0, 0, False, node_values, max_depth, alpha, beta)
    print(f"Optimal value for maximizer: {result_max}\n")


```

```python
max_depth = 3
node_values = [4,8,9,3,2,-2,9,-1]

run_alphabeta(max_depth, node_values)
```
